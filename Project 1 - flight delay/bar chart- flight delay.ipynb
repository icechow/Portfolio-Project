```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
flight_filepath = "./flight_delays.csv"
```

```python
flight_data = pd.read_csv(flight_filepath, index_col= "Month")
```

```python
flight_data
```

```python
# Set the width and height of the figure
plt.figure(figsize=(8,6))

plt.ylim(0,12)
plt.grid(True)

# Add title
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
# the "Month" column was used to index the rows. We always have to use this special notation to select the indexing column.
sns.barplot(x=flight_data.index, y=flight_data['OO'])

# Add label for vertical axis
plt.ylabel("Arrival delay (in minutes)")
```

```python
# Set the width and height of the figure
plt.figure(figsize=(15,7))

# Add title
plt.title("Average Arrival Delay for Each Airline, by Month")

# Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=flight_data, annot=True, linewidths= 0.4, linecolor="grey" )

# Add label for horizontal axis
plt.xlabel("Airline")
```
