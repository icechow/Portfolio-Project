```python
import random

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}
playing = True
```

```python
class Card:
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
        
    def __str__(self):
        return self.rank + ' of ' + self.suit


class Deck:
    
    def __init__(self):
        self.deck = []
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit,rank))
    
    def __str__(self):
        deck_comp = ''
        for card in self.deck:
            deck_comp += '\n' +card.__str__()
        return 'The deck has: ' + deck_comp
        
    def shuffle(self):
        random.shuffle(self.deck)
    
    def deal(self):
        return self.deck.pop()
    
```

```python
class Chips:
    
    def __init__(self):
        self.total = 100
        self.bet = 0
    
    def win_bet(self):
        self.total += self.bet
    
    def lose_bet(self):
        self.total -= self.bet
        
```

```python
def take_bet(chips):
    
    while True:
        
        try:
            chips.bet= int(input("How many chips you want to bet?"))
        
        except ValueError:
            print("Must be an initeger")
            
        else:
            if chips.bet > chips.total:
                print('Your bet can not exceed ',chips.total)
            else:
                break
```

```python
def hit(deck,hand):
    hand.add_card(deck.deal())
    hand.adjust_for_ace
    
```

```python
def hit_or_stand(deck,hand):
    
    global playing
    
    while True:
        
        x = input("Would you like to hit or stand? Enter 'h' or 's'.")
        
        if x[0].lower() == 'h':
            hit(deck,hand)
            
        elif x[0].lower() == 's':
            print ("Player stand, Dealing still playing.")
            playing = False
            
        else:
            print("Please input again.")
            continue
        break
            
```
